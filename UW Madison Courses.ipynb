# Effect of number of classes in a week

Here we are checking that is there any effect on grade of student based on how many classes scheduled for each subject within a week. For example: Math class: is twice a week, English class: Once a week (just an example, may not be true). So, we want to see there is any changes in grades if one class held 7 days a week as compare to 1 day a week.

### Steps:

1) merge schedule data with section to get tge course offering id.

2) merge the new data to grade dataset to get the grades.

3) filter the dataset where there is no class in whole week.

4) Group by number of class to see the impact on grades


```python
schedule_merge = schedules.merge(sections[['course_offering_uuid','schedule_uuid']], how = 'left',left_on='uuid',right_on= 'schedule_uuid')
grade_merge = grade_distributions[['course_offering_uuid', 'section_number', 'a_count', 'ab_count',
       'b_count', 'bc_count', 'c_count', 'd_count', 'f_count']].copy()
```

```python
test = schedule_merge.merge(grade_merge, how = 'left', on = 'course_offering_uuid')
#remove all the classes which are not scheduled 
test = test[~
((test['tues'] == False) &
(test['sun'] == False) &
(test['mon'] == False) &
(test['wed'] == False) &
(test['thurs'] == False) &
(test['fri'] == False) &
(test['sat'] == False)) ]
print(test.shape)
```

```python
test_rep = test.replace({True:1,False:0})
test_rep['no_cls'] = test_rep['mon'] +test['tues']+test['sun']+test['wed']+test['thurs']+test['fri']+test['sat']
test_rep = grade_avg(test_rep)
```

```python
# no significant results to say no of classes affect the grades
test_rep.groupby(by = test_rep['no_cls']).agg({'a_count':'mean',
'ab_count':'mean',
'b_count':'mean',
'bc_count':'mean',
'c_count':'mean',
'd_count':'mean',
'f_count':'mean',
'uuid':'count'})
```

As number of classes does not have significant effect on the grades we will see if classes or days i.e. Monday, Friday, Sunday affects the grades.
For example: class grade on Monday vs Sunday or weekday vs weekend.

For that purpose, we are only considering the classes which held only once a week. One can dig deep by comparing all the possible ways. 

# Effect of Week Day on grades.

Analyze the class grade based on the day class scheduled.

### Steps:
1) Filter the classes which only held once a week.

2) Group by according the weekday and average the grades to analyze the effects.

```python
days_affect = test_rep[test_rep['no_cls']==1].copy()
x = days_affect[['mon','tues','wed','thurs','fri','sat','sun']].idxmax(axis = 1)
days_affect['day'] =x
days_affect = days_affect[days_affect['a_count'].notnull()]
days_affect_grop = days_affect.groupby(by = days_affect['day']).agg({'a_count':'mean',
'ab_count':'mean',
'b_count':'mean',
'bc_count':'mean',
'c_count':'mean',
'd_count':'mean',
'f_count':'mean',
'uuid':'count'})
days_affect_grop
```

# Strict Instructor
We can help students by flagging each professor according their grading pattern. It will give students a preview of how strict a professor grades and then they can decide which course they want to take. Sometimes in universities there will be multiple professors teaching the same course, so this might students to select whose classes they want to take.

### steps:
1) Merge the grading dataset to section instructor to get the unique id for each instructor.

2) Merge new dataset with the instructorâ€™s dataset to get the name. (you do not need to this step if you don't want to the names)

3) calculate the % of people for each grade.

4) At last average the grades by individual processor to identify how strict is the checking.

5) Sort it by Fail, D grade and so on.


```python
Grade_section = grade_distributions.merge(sections, how = 'left',on = 'course_offering_uuid')
grade_prof = Grade_section.merge(teachings, how = 'left',left_on = 'uuid', right_on = 'section_uuid')
strict_instructor = grade_prof[['course_offering_uuid','a_count', 'ab_count',
       'b_count', 'bc_count', 'c_count', 'd_count', 'f_count', 'instructor_id']].copy()
strict_instructor = grade_avg(strict_instructor)
```

```python
strict_instructor = strict_instructor.merge(instructors,how = 'left', left_on='instructor_id',right_on= 'id')
strict_instructor.columns
```

```python
strict_instructor = strict_instructor.groupby(by = strict_instructor['name']).agg({'a_count':'mean',
'ab_count':'mean',
'b_count':'mean',
'bc_count':'mean',
'c_count':'mean',
'd_count':'mean',
'f_count':'mean',
'course_offering_uuid':'count'}).sort_values(['f_count','d_count',
                                                           'c_count','b_count'],ascending=False)
```
